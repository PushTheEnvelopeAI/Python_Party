# 2. Functions

```python
def add_function(num1, num2):
    num3 = num1 + num2
    return num3

def print_sum(num1, num2, num3):
    print('%d + %d = %d'%(num1, num2, num3))

    
num1s = [1,4,2,5,7]
num2s = [3,6,1,6,8]

for num1, num2 in zip(num1s, num2s):
    num3 = add_function(num1, num2)
    print_sum(num1, num2, num3)
```

# EXERCISE 2: Create a function to convert a given GPA to a letter grade
1. Make a list with the following GPA values: [ 2.6, 4.0, 1.7, 3.1, 3.9, 1.1, 3.4, 0.2 ]
2. Build a function to take in a GPA value and return a letter grade
- 4.00: A
- 3.67: A-
- 3.33: B+
- 3.00: B
- 2.67: B-
- 2.33: C+
- 2.00: C
- 1.67: C-
- 1.33: D+
- 1.00: D
- 0.67: D-
- 0.00: F
3. Build a function to print the GPA grade and the converted letter grade

```python
def convert_grade(grade):
    if grade == 4.00:
        letter = 'A'
    elif grade >= 3.67 and grade < 4.0:
        letter = 'A-'
    elif grade >= 3.33 and grade < 3.67:
        letter = 'B+'
    elif grade >= 3.00 and grade < 3.33:
        letter = 'B'
    elif grade >= 2.67 and grade < 3.00:
        letter = 'B-'
    elif grade >= 2.33 and grade < 2.67:
        letter = 'C+'
    elif grade >= 2.00 and grade < 2.33:
        letter = 'C'
    elif grade >= 1.67 and grade < 2.00:
        letter = 'C-'
    elif grade >= 1.33 and grade < 1.67:
        letter = 'D+'
    elif grade >= 1.00 and grade < 1.33:
        letter = 'D'
    elif grade >= 0.67 and grade < 1.00:
        letter = 'D-'
    elif grade < 0.67:
        letter = 'F'
    else:
        letter = 'error'
        
    return letter

def print_grade(grade, letter):
    print('GPA: %1.2f | Letter Grade: %s'%(grade, letter)) 
        
        
grades = [2.6, 4.0, 1.7, 3.1, 3.9, 1.1, 3.4, 0.2]

for grade in grades:
    letter = convert_grade(grade)
    print_grade(grade, letter)
```
