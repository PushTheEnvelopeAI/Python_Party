### Import Pyplot

```python
import matplotlib.pyplot as plt
```

### Read File using Pandas

```python
import pandas as pd

data = pd.read_csv("files/stock_data.txt", delimiter=",")
data['Date'] = pd.to_datetime(data['Date'])
data['Change'] = data['Close'] - data['Open']
data['Bin'] = pd.cut(data['Change'], bins=[-200, -2, 2, 200],  labels=['Negative', 'None', 'Positive'])

display(data)
```

### Line Graphs

```python
plt.figure(figsize=(15,7))
plt.plot(data['Date'], data['Close'], c="red")
plt.plot(data['Date'], data['Open'], c="blue")
plt.title("TSLA Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

### Scatter Plots

```python
plt.figure(figsize=(15,7))
plt.scatter(data['Date'], data['Close'], c="red", s=5)
plt.scatter(data['Date'], data['Open'], c="blue", s=5)
plt.title("TSLA Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

### Histograms

```python
plt.figure(figsize=(15,7))
plt.hist(data['Change'], bins=40)
plt.title("Price Changes")
plt.xlabel("Date")
plt.ylabel("Price Changes")
plt.show()
```

### Pie Charts

```python
plt.figure(figsize=(12,12))
plt.pie(data['Bin'].value_counts().values, labels = data['Bin'].value_counts().index.values)
plt.title("Ratio of Percent Changes")
plt.show()
```
