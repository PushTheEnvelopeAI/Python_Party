### Import Pandas

```python
import pandas
import pandas as pd
from pandas import read_csv
```

### Read File

```python
stockFile = pd.read_csv("files/stock_data.txt", delimiter=",")
display(stockFile)

shareFile = pd.read_csv("files/my_shares.txt", delimiter=",")
display(shareFile)
```

### Add/Edit Columns

```python
stockFile['Change'] = stockFile['Close'] - stockFile['Open']
display(stockFile)
```

### Combining Files

```python
combined = stockFile.merge(shareFile, on="Date", how='left')
combined['Diffs'] = combined['Change'] * combined['Shares']
display(combined)
```

### Write Files

```python
combined = combined[['Diffs']]
combined.to_csv("files/my_account.txt", index=False, header=False)
```

### Column Statistics

```python
print(1000-combined['Diffs'].sum())
```

### Filtering Files

```python
display(stockFile[stockFile['Change']>10])
```
